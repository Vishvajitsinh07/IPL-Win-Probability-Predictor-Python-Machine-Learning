```python
teams=[
    'Chennai Super Kings',
    'Mumbai Indians',
    'Kolkata Knight Riders',
    'Rajasthan Royals',
    'Sunrisers Hyderabad',
    'Delhi Capitals',
    'Lucknow Super Giants',
    'Gujarat Titans',
    'Royal Challengers Bengaluru',
    'Punjab Kings'
]
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

match_df['team1'] = match_df['team1'].str.replace('Royal Challengers Bangalore','Royal Challengers Bengaluru')
match_df['team2'] = match_df['team2'].str.replace('Royal Challengers Bangalore','Royal Challengers Bengaluru')

match_df['team1'] = match_df['team1'].str.replace('Kings XI Punjab','Punjab Kings')
match_df['team2'] = match_df['team2'].str.replace('Kings XI Punjab','Punjab Kings')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
match_df = match_df[['match_id','city','winner','total_runs']]
```

```python
match_df.merge(delivery,on='match_id')
```

```python
delivery_df = match_df.merge(delivery,on='match_id')
```

```python
delivery_df = delivery_df[delivery_df ['inning']==2]
```

```python
delivery_df
```

```python
delivery_df.shape
```

```python
delivery_df
```

```python
delivery_df['current_score'] = delivery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery_df
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']+1
```

```python
delivery_df
```

```python
delivery_df['balls_left'] = 120 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df
```

```python
wickets = delivery_df.groupby('match_id')['is_wicket'].cumsum().values
delivery_df['wickets_left'] = 10 - wickets
delivery_df.head()
```

```python
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
delivery_df
```

```python
final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets_left','total_runs_x','crr','rrr','result']]
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df
```

```python
final_df.dropna(inplace=True)
```

```python
final_df = final_df[final_df['balls_left'] != 0]
```

```python
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=200)
```

```python
import joblib
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression

column_transformer = ColumnTransformer(
    [('trf', OneHotEncoder(sparse=False, drop='first'), ['batting_team', 'bowling_team', 'city'])],
    remainder='passthrough'
)

pipeline = Pipeline([
    ('step1', column_transformer), 
    ('step2', LogisticRegression(solver='liblinear'))  
])

pipeline.fit(X_train,y_train)

```

```python

joblib.dump(pipeline, 'ipl_win_probability_model_final.pkl')
```

```python
y_pred=pipeline.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
import joblib
import numpy as np
import pandas as pd


model = joblib.load('ipl_win_probability_model_final.pkl')


teams_full = {
    'CSK': 'Chennai Super Kings',
    'MI': 'Mumbai Indians',
    'KKR': 'Kolkata Knight Riders',
    'RR': 'Rajasthan Royals',
    'SRH': 'Sunrisers Hyderabad',
    'DC': 'Delhi Capitals',
    'LSG': 'Lucknow Super Giants',
    'GT': 'Gujarat Titans',
    'RCB': 'Royal Challengers Bengaluru',
    'PBKS': 'Punjab Kings'
}

cities = [
    'Mumbai', 'Chennai', 'Kolkata', 'Bangalore', 'Delhi', 'Hyderabad', 
    'Jaipur', 'Ahmedabad', 'Pune', 'Visakhapatnam', 'Raipur', 'Indore', 
    'Mohali', 'Nagpur', 'Kanpur', 'Dharamsala', 'Rajkot', 'Thiruvananthapuram', 
    'Lucknow', 'Cuttack'
]


def get_full_team_name(team_name):
    team_name = team_name.upper()
    for shortcut, full_name in teams_full.items():
        if team_name == shortcut or team_name == full_name.upper():
            return full_name
    return None


def predict_win_probability():
    # Display teams
    print("Available teams:")
    for shortcut, full_name in teams_full.items():
        print(f"{shortcut}: {full_name}")
    
    
    while True:
        batting_team = input("Enter Batting Team: ")
        batting_team = get_full_team_name(batting_team)
        if not batting_team:
            print("Invalid team name. Please try again.")
        else:
            break
    
   
    while True:
        bowling_team = input("Enter Bowling Team: ")
        bowling_team = get_full_team_name(bowling_team)
        if not bowling_team:
            print("Invalid team name. Please try again.")
        elif bowling_team == batting_team:
            print("Bowling team cannot be the same as batting team. Please try again.")
        else:
            break
    
    
    print("\nAvailable cities for the venue:")
    for city in cities:
        print(city)
    
    
    while True:
        city = input("\nEnter city: ").capitalize()
        if city not in cities:
            print("Invalid city. Please try again.")
        else:
            break

    
    while True:
        try:
            wickets_left = int(input("Enter wickets left (1-10): "))
            if 0 < wickets_left <= 10:
                break
            else:
                print("Invalid number of wickets. Please enter a value between 1 and 10.")
        except ValueError:
            print("Invalid input. Please enter an integer value.")
    
    
    while True:
        try:
            overs_left = float(input("Enter overs left (0-20): "))
            if 0 < overs_left <= 20:
                balls_left = int(overs_left * 6) 
                break
            else:
                print("Invalid number of overs. Please enter a value between 0 and 20.")
        except ValueError:
            print("Invalid input. Please enter a numeric value.")
    
    
    while True:
        try:
            current_score = int(input("Enter current score: "))
            if current_score >= 0:
                break
            else:
                print("Invalid score. Please enter a positive integer.")
        except ValueError:
            print("Invalid input. Please enter an integer value.")
    
    
    while True:
        try:
            target = int(input("Enter target score: "))
            if target > current_score:
                runs_left = target - current_score
                break
            else:
                print("Target score must be greater than current score. Please try again.")
        except ValueError:
            print("Invalid input. Please enter an integer value.")

   
    if balls_left > 0 and balls_left < 120:
        crr = (current_score * 6) / (120 - balls_left)
        rrr = (runs_left * 6) / balls_left
    elif balls_left == 120:
        crr = 0
        rrr = (runs_left * 6) / balls_left
    else:
        
        crr = 0
        rrr = float('inf')
    
    
    input_data = pd.DataFrame({
        'batting_team': [batting_team],
        'bowling_team': [bowling_team],
        'city': [city],
        'runs_left': [runs_left],
        'balls_left': [balls_left],
        'wickets_left': [wickets_left],
        'total_runs_x': [target],
        'crr': [crr],
        'rrr': [rrr]
    })

    win_prob = model.predict_proba(input_data)[0]
    

    print(f"\nWin Probability for {batting_team}: {win_prob[1] * 100:.2f}%")
    print(f"Win Probability for {bowling_team}: {win_prob[0] * 100:.2f}%")

predict_win_probability()
```
